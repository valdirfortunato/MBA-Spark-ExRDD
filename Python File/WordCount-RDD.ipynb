## Import PySpark

```python
import findspark
from pyspark.sql import SparkSession
```

## Create SparkSession

```python
findspark.init()
from pyspark import SparkContext, SparkConf
sc = SparkContext(appName="Word Count")
```

## Read file

```python
rdd = sc.textFile(r"C:\Users\valdi\Documents\Spark\Exercicio-RDD\Source\lorem_ipsum.txt")
print(rdd.collect())
```

## Word Count

```python
# Efetuando a contagem das palavras
counts = rdd.flatMap(lambda line: line.split(" ")) \
                            .map(lambda word: (word, 1)) \
                           .reduceByKey(lambda x, y: x + y)

# Exibindo 
output = counts.collect()
for (word, count) in output:
    print("%s: %i" % (word, count))
```

## Stopping session

```python
# Parando a sess√£o do Spark
sc.stop()
```
